# Author stats search

```python
#number of papers per year histogram by an author

from numpy.ma.core import left_shift
from textwrap import fill
from pandas._libs.algos import backfill
from re import search
import plotly.graph_objects as go
app = JupyterDash(__name__)
colors = {
    'background': '#111111',
    'text': '#7FDBFF'
}
fig = go.Figure(data=[go.Table(
        columnwidth = [300,150,150,150],
        header=dict(values=['Author', 'Number of Citations', 'hIndex','Number of Publications'],
                    fill_color='grey',
                    font=dict(color='white', size=12),
                    align='left'),
        cells=dict(fill_color='white',
                  align='left'))
        ])
fig2 = px.bar(df, x="Year", y="paperCount",
                barmode="group",
                labels=dict(Year="Year", paperCount="Number of Publications"))
fig.update_layout(
        font_family="Courier New",
        autosize=False,
        width=1225,
        height=100,
        margin=dict(l=0, r=0, t=0, b=0),
    )
fig2.update_layout(
        font_family="Courier New",
        autosize=False,
        width=1225,
        height=400,
        margin=dict(l=0, r=0, t=0, b=0),
    )
fig2.update_xaxes(range=[1950,2025])
app.layout = html.Div(
    style = {
        "color":"Black",
        "background-color":"white",
        "font-family":"Courier New",
        "width": "95%",
        "padding": "10px 10px",
        "border-radius": "10px",
    },
    children=[
    html.H2("Number of Publications", style = {"font-weight":"normal",}),
    html.Label([
        dcc.Dropdown(
            id='csv-dropdown', clearable=False,
            value='Select an Author...', options=[
                {'label': c, 'value': c}
                for c in df3['FirstLast']
            ])
    ]),
    dcc.Graph(figure=fig,id='table',
        style = {
            "left":"0px",
            "top":"0px",
        }
    ),
    dcc.Graph(figure=fig2,id='table2',
        style = {
            "position":"relative",
        }
    ),
])
# Define callback to update graph
@app.callback(
    Output('table', 'figure'),
    Output('table2', 'figure'),
    [Input("csv-dropdown", "value")]
)


# Dropdown to choose the Author
# Generate table with CitationCount, HIndex
# Generate chart with Number of Papers per Year
def generate_table(authorselection):
    dfsearch = df3[df3['FirstLast'] == authorselection]
    fig = go.Figure(data=[go.Table(
        columnwidth = [300,150,150,150],
        header=dict(values=['Author', 'Number of Citations', 'hIndex','Number of Publications'],
                    fill_color='grey',
                    font=dict(color='white', size=12),
                    align='left'),
        cells=dict(values=[dfsearch.FirstLast, dfsearch.CitationCount, dfsearch.hIndex, dfsearch.paperCount],
                  fill_color='white',
                  align='left'))
        ])
    dfsearch1 = df[df['FirstLast'] == authorselection]
    dfpaperCount = df3[df3['FirstLast'] == authorselection]
    temp = dfpaperCount['paperCount']
    dfsearch1['paperCount'].replace({temp.iloc[0]: 1}, inplace=True)
    fig2 = px.bar(dfsearch1, x="Year", y="paperCount",
                barmode="group",
                labels=dict(Year="Year", paperCount="Number of Publications"))
    fig.update_layout(
        font_family="Courier New",
        autosize=False,
        width=1225,
        height=100,
        margin=dict(l=0, r=0, t=0, b=0),
    )
    fig2.update_layout(
        font_family="Courier New",
        autosize=False,
        width=1225,
        height=400,
        margin=dict(l=0, r=0, t=0, b=0),
    )
    fig2.update_xaxes(range=[1950,2025])
    return fig, fig2

# Run app and display result inline in the notebook
app.run_server(mode='inline')
```
